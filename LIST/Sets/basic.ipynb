subset

```python
#checks if all element of one set are in another
a={1,2,3,4,5,6}
b={6,8,9,10,11}
a.issubset(b)
```

issuperset

```python
#check if one set contain all element in comman
a={1,2,3,4,5,6}
b={6,8,9,10,11}
a.isdisjoint(b)
```

isdisjoint

```python
#return true if sets have no elemnet in comman

a.isdisjoint(b)
```

intersection update

```python
a={1,2,3}
b={2,3,4}
#update the set to keeps only item found in both sets
a.intersection_update(b)
print(a)
```

difference_update

```python
#remove element found in another set from the current set

a.difference_update(b)
print(a)
```

symmetric_update

```python
#update the set to the symmetric difference of itself and another

a.symmetric_difference_update(b)
print(a)
```
