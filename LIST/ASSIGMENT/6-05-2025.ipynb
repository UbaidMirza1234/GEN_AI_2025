The if-elif-else statement in Python is used for decision-making and allows checking multiple conditions. The flow of control follows these rules:

- The program evaluates the if condition first. If it’s true, the corresponding block executes, and the rest of the conditions are skipped.
- If the if condition is false, the program checks the elif condition(s) sequentially. The first true elif condition gets executed.
- If none of the conditions are met, the else block executes.

2.  What is the difference between `for` loops and `while` loops in terms of conditional checks? When would you prefer one over the other? 
   -  Coding Challenge:  Write a Python script to print all prime numbers between 1 and 100 using a `for` loop.


- for loop:
- Iterates over a sequence (e.g., list, range, string).
- The number of iterations is known before execution.
- Used when you need to iterate through elements systematically.

- while loop:
- Runs as long as a specified condition is True.
- Number of iterations is not known beforehand.
- Used when looping should continue based on a dynamic condition.



```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Print all prime numbers between 1 and 100
for num in range(1, 101):
    if is_prime(num):
        print(num, end=" ")
```

3.  Explain how nested `if` statements work in Python. How can you avoid deep nesting to make your code more readable? 
   -  Coding Challenge:  Write a function that takes three numbers as input and returns the largest of the three using nested `if` statements.


Nested if statements occur when an if statement is placed inside another if statement. This allows more granular conditional checks, but too much nesting can make code harder to read.


How to Avoid Deep Nesting?
- Use logical operators (and, or) to combine conditions.
- Return early to simplify conditional branching.
- Leverage elif to avoid unnecessary nested conditions.
- Refactor into separate functions if logic is complex.


```python
x = 10
y = 20

if x > 0:
    if y > 10:
        print("Both x and y are positive and y is greater than 10")
```

```python
def find_largest(a, b, c):
    if a > b:
        if a > c:
            return a
        else:
            return c
    else:
        if b > c:
            return b
        else:
            return c

# Example usage:
print(find_largest(10, 25, 15))  # Output: 25
```

4.  What are the potential risks of using `break` in loops? How does it affect loop execution and what alternatives can you use? 
   -  Coding Challenge:  Write a Python program that reads numbers from the user until they input a negative number. The program should then print the sum of all positive numbers entered.

Risks of Using break in Loops
- Abrupt Termination: Stops the loop immediately, skipping remaining iterations.
- Reduced Readability: Overuse makes logic harder to follow.
- Loss of Expected Behavior: Can disrupt structured flow, leading to unexpected results.
- Skipping Cleanup Operations: If handling resources, premature exit may bypass important steps.

Alternatives to break
- Boolean Flags: Use a variable to track conditions instead of breaking.
- Modify Loop Conditions: Adjust the loop's stopping condition dynamically.
- Use return in Functions: If inside a function, return early instead of breaking.
Want a quick example for any of these approaches?


```python
def sum_positive_numbers():
    total = 0

    while True:
        num = int(input("Enter a number (negative to stop): "))
        if num < 0:
            break  # Exit the loop when a negative number is entered
        total += num  

    print(f"Sum of positive numbers: {total}")

# Run the program
sum_positive_numbers()
```

q5) How does the else clause in a loop work, and how does it differ from the else in conditional statements?
- Coding Challenge:  Write a function that searches for a specific element in a list using a for loop. If the element is found, return its index; if not, return -1 using the else clause in the loop.


Understanding the else Clause in Loops
In Python, both for and while loops can have an else clause, which runs only if the loop completes fully without encountering a break statement.
Key Difference Between else in Loops vs. Conditionals
- In conditional statements (if-else), the else block runs if the if condition is false.
- In loops (for-else or while-else), the else block runs only if the loop completes without breaking.


```python
def search_element(lst, target):
    for index, value in enumerate(lst):
        if value == target:
            return index  
    else:
        return -1 

# Example usage
numbers = [10, 20, 30, 40, 50]
print(search_element(numbers, 30))  
print(search_element(numbers, 60))  
```

6.  What is a common pitfall when using floating-point numbers in conditional statements? How can you avoid it? 
   -  Coding Challenge:  Write a Python function that compares two floating-point numbers and returns `True` if they are approximately equal, considering a small tolerance value.


Common Pitfall with Floating-Point Numbers in Conditional Statements
Floating-point arithmetic in Python (and most programming languages) can introduce precision errors due to the way numbers are stored in binary format. This often causes unexpected results in comparisons.




```python
import math

def are_almost_equal(x, y, tolerance=1e-9):
    return math.isclose(x, y, abs_tol=tolerance)

# Example usage
print(are_almost_equal(0.1 + 0.2, 0.3))  # Output: True
```

7.  How can you combine `for` loops and `if` statements to filter and process data in Python? 
   -  Coding Challenge:  Given a list of integers, write a Python program that uses a `for` loop and `if` statements to create a new list containing only the even numbers.


Combining for Loops and if Statements for Data Processing
Python's for loops are great for iterating over lists, while if statements allow filtering data by specific conditions. Together, they help process and extract meaningful data efficiently


```python
def filter_even_numbers(lst):
    even_lst = [num for num in lst if num % 2 == 0]  # List comprehension method
    return even_lst

# Example usage
numbers = list(range(1, 21))  # Generate numbers 1 to 20
print(filter_even_numbers(numbers))  # Output: [2, 4, 6, ..., 20]
```

Q8) Discuss the concept of short-circuit evaluation in Python. How does it affect the performance of conditional statements?
- Coding Challenge:  Write a function that takes three boolean values and returns True if at least two of them are True, using short-circuit evaluation.


Understanding Short-Circuit Evaluation in Python
Short-circuit evaluation is a technique where Python stops evaluating logical expressions as soon as the outcome is determined.

1) For or statements:
- If any operand is True, Python immediately returns True without checking the remaining conditions.

2) For and statements:
- If any operand is False, Python immediately returns False without checking further.

```python
def at_least_two_true(a, b, c):
    return (a and b) or (a and c) or (b and c)

# Example usage:
print(at_least_two_true(True, False, True))  # Output: True
print(at_least_two_true(False, False, True)) # Output: False
```

9.  Explain how the `continue` statement works in a loop. What are some scenarios where using `continue` is more beneficial than restructuring the loop? 


   -  Coding Challenge:  Write a Python program that iterates through a list of numbers and prints only those numbers that are divisible by 3, using the `continue` statement.

Understanding the continue Statement in Loops
The continue statement is used inside loops to skip the remaining code in the current iteration and move to the next iteration without terminating the loop.


When Is continue More Beneficial Than Restructuring?
- Skipping Specific Cases Efficiently: Instead of using multiple conditional checks or restructuring with nested loops, continue cleanly skips undesired iterations.
- Filtering Data in Loops: When processing a list, continue is useful to ignore invalid or unwanted elements while keeping the loop running.
- Avoiding Complex Nested Conditions: If skipping an iteration is the main goal, continue simplifies code instead of using deeply nested if statements


```python
numbers = list(range(1, 21))  # Creating a list of numbers from 1 to 20

for num in numbers:
    if num % 3 != 0:  
        continue
    print(num)  

```

10.  How can you use list comprehensions in combination with conditional statements to make your code more concise? 
    -  Coding Challenge:  Write a list comprehension that generates a list of squares of all even numbers between 1 and 20.


```python
Using List Comprehensions with Conditional Statements
List comprehensions provide a compact and readable way to create lists. When combined with conditional statements, they allow filtering and processing data in a single line, improving efficiency
```

```python
# Generating squares of all even numbers between 1 and 20
squares_of_even_numbers = [num**2 for num in range(1, 21) if num % 2 == 0]

print(squares_of_even_numbers)
```
