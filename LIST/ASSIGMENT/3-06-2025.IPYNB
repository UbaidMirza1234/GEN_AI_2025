2) Explain how Python handles type conversion between different data types, such as between integers and floats or between strings and lists.

```python

ANS  Python handles type conversion in two ways:
    1. Implicit Type Conversion (Automatic)
    2.EXPLICIT  Type Conversion (Type Casting)

1. Implicit
          Python automatically converts one data type to another when safe to do so (usually from a lower to higher data type).

a = 5       # int
b = 2.5     # float
c = a + b   # int + float → float
print(c)    # Output: 7.5


2. EXPLICIT  
           You manually convert data from one type to another using built-in functions.


| Conversion Type | Function        | Example                      |
| --------------- | --------------- | ---------------------------- |
| Int to Float    | `float(x)`      | `float(5)` → `5.0`           |
| Float to Int    | `int(x)`        | `int(3.9)` → `3` (truncates) |
| String to Int   | `int("10")`     | `"10"` → `10`                |
| String to Float | `float("3.14")` | `"3.14"` → `3.14`            |
| Int to String   | `str(100)`      | `100` → `"100"`              |
| List to String  | `str([1, 2])`   | `[1, 2]` → `"[1, 2]"`        |
| String to List  | `list("abc")`   | `"abc"` → `['a', 'b', 'c']`  |
```

4 .Discuss the role of the `__repr__` and `__str__` methods in custom data types. How do they differ, and when should you implement them?

```python
1. __repr__(self)
Goal: Unambiguous string representation (used for debugging and development).

Should ideally return a string that can be used to recreate the object.

Called by:

1 repr(obj)

2 Interactive console (>>> obj)

3 Default fallback for str(obj) if __str__ is not defined.

 2. __str__(self)
Goal: Readable or user-friendly string representation.

Called by:

1 print(obj)

2 str(obj)


When to Implement:

Always implement __repr__ for debugging and logging.

Implement __str__ when you want a custom, readable printout for users.
```

6 What is the difference between the `+=` operator and the `+` operator in Python when used with mutable and immutable types?

```python


| Type      | Operator | Behavior          | New Object Created?  |
| --------- | -------- | ----------------- | -------------------  |
| Immutable | `+`      | Concatenates      | ✅ Yes               |
| Immutable | `+=`     | Rebinds           | ✅ Yes               |
| Mutable   | `+`      | Concatenates      | ✅ Yes               |
| Mutable   | `+=`     | Modifies in-place | ❌ No (in-place)     |

Conclusion:
Use + when you want to create a new object.

Use += when working with mutable types and want to modify in place (more memory-efficient).

With immutable types, += is essentially the same as + followed by assignment.
```

7  Explain the purpose and use of the `in` operator in Python. How does it behave differently when used with different data types, such as strings, lists, and dictionaries?

```python
Purpose and Use of the in Operator in Python
The in operator is used to check for membership — i.e., to test whether a value exists inside a container like a string, list, tuple, set, or dictionary.

| Data Type      | Checks For...                 | Example                      | Result |
| -------------- | ----------------------------- | ---------------------------- | ------ |
| **String**     | Substring                     | `'py' in 'python'`           | `True` |
| **List/Tuple** | Element                       | `3 in [1, 2, 3, 4]`          | `True` |
| **Set**        | Element                       | `2 in {1, 2, 3}`             | `True` |
| **Dictionary** |  **Keys only** (not values)   | `'name' in {'name': 'John'}` | `True` |
```
