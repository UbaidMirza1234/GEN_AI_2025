```python
#map -----function iterable
def sq(n):
    return n*n
```

```python
l = [2, 3, 4, 5]
Result = list(map(lambda x: x * x, l))
print(Result)
```

```python
#filter
l = [2, 3, 4, 5]
Result = list(filter(lambda x: x % 2 ==0, l))
print(Result)
```

```python
#Reduce -----function,iterable) iterable ussko ek single item
#left to right
from functools import reduce

l = [2, 3, 4, 5]
Result = reduce(lambda x, y: x + y, l)
print(Result)
#[5,4,5]------[9.5]------[14]
```

```python
#enumerate   --------iterable,start=0
l = [2, 3, 4, 5]
Result = list(enumerate(l))
print(Result)
```
