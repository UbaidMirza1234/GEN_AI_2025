```python
#factorial
n=int(input("enter the number"))

def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Input from the user
num = int(input("Enter a number: "))
print(f"The factorial of {num} is {factorial(num)}")
```
